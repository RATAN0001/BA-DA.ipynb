```python
import openpyxl
import pandas as pd
workbook = openpyxl.Workbook() #Opening Excelsheet
sheet =workbook.active   #activating the worksheet
data =[
    ['Name','DOMAIN','AGE','LOCATION','SALARY','EXP'],
    ['John','Testing',26,'Bangalore',8000,1],
    ['Khushi','Java',30,'Hyderabad',15000,2],
    ['Charlie','C#',33,'Delhi',30000,4],
    ['Deepa','DataAnalyst',24,'Bangalore',40000,5],
    ['Sahana','DataEngineer',35,'Mumbai',60000,8]

]
for row in data:
  sheet.append(row)
workbook.save('data.xlsx')
emp = pd.read_excel('data.xlsx')
print(emp)
```
